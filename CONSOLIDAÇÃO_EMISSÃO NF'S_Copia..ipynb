```python
import pandas as pd
import py
#pôr os cnpjs sem barras, pontos e hífens, pois no sistema do compliance é necessário.
def corrigir_cnpjs(cnpj):
    cnpj = cnpj.replace('/','').replace('-','').replace('.','')
    return cnpj
#ignorar possíveis espaços no dataframe que podem interferir na descrição da NF.
def corrigir_tabs(tabs):
    tabs = tabs.replace('\t',' ').replace('\n',' ')
    return tabs

#formatar valores por milhar de acordo com o sistema do compliance.
def formatar(valor):
    return "{:,.2f}".format(valor)



tabela = pd.read_excel("BASE DE DADOS.NF.TESTE.Copia.xlsx")

display(tabela)#antes da funções de formatação

tabela['CNPJ'] = tabela['CNPJ'].astype(str)

tabela['CNPJ'] = tabela['CNPJ'].apply(corrigir_cnpjs)

tabela['VALOR'] = tabela['VALOR'].apply(formatar)

tabela['DESCRIÇÃO NF'] = tabela['DESCRIÇÃO NF'].apply(corrigir_tabs)#ESSA É A NOVA FUNÇÃO P TIRAR O \T Q AS VEZES VEM.


display(tabela)

print('-' * 15)

print(tabela.info())
```
