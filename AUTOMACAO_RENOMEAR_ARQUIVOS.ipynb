```python
import pandas as pd

def corrigir_zeros(dado):
    dado = dado.replace('.0','')
    return dado

df = pd.read_excel(r'C:/Users/vinicius.dossantos/OneDrive - BRMALLS/Área de Trabalho/ARQUIVOS_CG_FPP_MONTANTE/CARGA_RENOMEAR_ARQUIVOS - Copia.xlsx')
df['TRANSAÇÃO'] = df['TRANSAÇÃO'].astype(str)
df['NF'] = df['NF'].astype(str)
df['NF'] = df['NF'].apply(corrigir_zeros)
df['TRANSAÇÃO'] = df['TRANSAÇÃO'].apply(corrigir_zeros)



display(df)
df.info()
dicionario_01_NF = {}
dicionario_02_NF = {}
dicionario_03_NF = {}
dicionario_01_BOLETO = {}
dicionario_02_BOLETO = {}
dicionario_03_BOLETO = {}

for i, transacao in enumerate(df['TRANSAÇÃO']):
    sigla = df.loc[i, 'SIGLA']
    nf = df.loc[i, 'NF']
    pasta = df.loc[i,'PASTA']
    if pasta == 'BRM01':
        texto_nf = 'NF '
        texto_boleto = 'BOLETO '
        texto_2 = ' TRANSACAO'
        if 'CG' in sigla:
            texto_3 = ' TAXA ADM CG - '
        elif 'FPP' in sigla:
            texto_3 = ' TAXA ADM FPP - '
        texto_final_nf = texto_nf + nf + texto_2 + ' ' + transacao + texto_3 + sigla
        texto_final_boleto = texto_boleto + transacao + ' -' + texto_3 + sigla
        dicionario_01_BOLETO[transacao] = texto_final_boleto
        dicionario_01_NF[transacao] = texto_final_nf
    elif pasta == 'BRM02':
        texto_nf = 'NF '
        texto_boleto = 'BOLETO '
        texto_2 = ' TRANSACAO'
        if 'CG' in sigla:
            texto_3 = ' TAXA ADM CG - '
        elif 'FPP' in sigla:
            texto_3 = ' TAXA ADM FPP - '
        texto_final_nf = texto_nf + nf + texto_2 + ' ' + transacao + texto_3 + sigla
        texto_final_boleto = texto_boleto + transacao + ' -' + texto_3 + sigla
        dicionario_02_BOLETO[transacao] = texto_final_boleto
        dicionario_02_NF[transacao] = texto_final_nf
    elif pasta == 'BRM03':
        texto_nf = 'NF '
        texto_boleto = 'BOLETO '
        texto_2 = ' TRANSACAO'
        if 'CG' in sigla:
            texto_3 = ' TAXA ADM CG - '
        elif 'FPP' in sigla:
            texto_3 = ' TAXA ADM FPP - '
        texto_final_nf = texto_nf + nf + texto_2 + ' ' + transacao + texto_3 + sigla
        texto_final_boleto = texto_boleto + transacao + ' -' + texto_3 + sigla
        dicionario_03_BOLETO[transacao] = texto_final_boleto
        dicionario_03_NF[transacao] = texto_final_nf

    
print(dicionario_01_NF)
   
```
